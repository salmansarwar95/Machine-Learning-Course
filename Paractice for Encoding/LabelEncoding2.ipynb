```python
import pandas as pd
```

```python
datas = pd.read_csv("Car details v3.csv")
print(datas.isnull().sum())
datas.head(5)
```

```python
nd = (datas["fuel"])
(nd).unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
(datas["en_fuel"]) = le.fit_transform(nd)
(datas["en_fuel"]).unique()
```
