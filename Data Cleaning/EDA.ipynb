```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("random_data.csv")
print (df)
print("\n", df.count())
```

```python
nam = (df["Name"])
print("Number of Total Participants:\n", nam.count())

sns.countplot(x=nam, data=df, color="skyblue", edgecolor="black")
plt.title("Total Participants", fontsize="15")
plt.tight_layout()
plt.show()
```

```python
sx = (df["Sex"])

plt.figure(figsize=(4,5))
sns.countplot(x=sx, data=df, color="pink", edgecolor="black")
plt.title("Participants Genders", fontsize="15")
plt.tight_layout()
plt.show()
```

```python
ag = (df["Age"])

sns.countplot(x=ag, data=df, color="gray", edgecolor="black")
plt.title("Participants Ages", fontsize="15")
plt.tight_layout()
plt.show()
```

```python
mared = (df["Married"])

plt.figure(figsize=(4,5))
sns.countplot(x=mared, data=df, color="yellow", edgecolor="black")
plt.title("Marrital Status", fontsize="15")
plt.tight_layout()
plt.show()
```

```python
ocption = (df['Occupation'])

sns.countplot(x=ocption, data=df, color="green", edgecolor="black")
plt.title("Participants Occupation", fontsize="15")
plt.tight_layout()
plt.show()
```

```python
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.histplot(df["Age"], bins=10)
plt.title("Distribution of Total Bill")
plt.show()
```
