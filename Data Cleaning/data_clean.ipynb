```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("random_data.csv")
print(df.count())
print("\nData types:")
print(df.dtypes)
```

```python
ag = (df['Age'])
avg_age = np.mean(ag)
print(avg_age)

print("\n")

occ = (df["Occupation"])
print(occ.isnull().sum())
sns.countplot(x=occ, data=df, color = "pink", edgecolor = "black")
plt.title("Average Occupation")
plt.tight_layout()
plt.show()
```

```python
# for simple nul values
print(df.isnull().sum())
print(df.isnull().sum().sum())

#for percentae nul values
print((df.isnull().sum())/len(df)*100)
print((df.isnull().sum().sum())/len(df)*100)

print(df.count().sum())

plt.figure(figsize=(10,7))
sns.heatmap(df.isnull())
plt.show()
```
