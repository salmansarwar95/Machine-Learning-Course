```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("family.csv")
print ("Family Cousins:\n",df)
```

```python
age = (df["Age"])
Average = np.average(age)
print("Average Age:\n",Average)
```

```python
sns.countplot(x=age, data=df, color="Yellow", edgecolor="black")
plt.title("Ages Counting")
plt.tight_layout()
# plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x="Name", y="Age", data=df, color="pink", edgecolor="black")
plt.title("Family Cousins")
plt.xlabel("\nNames")
plt.ylabel("\nAges")
plt.tight_layout()
# plt.grid(True)
plt.show()
```
