```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Car details v3.csv")
print(dataset.head(2))
```

```python
print(dataset.isnull().sum())
```

```python
print(dataset.select_dtypes(include="object").isnull().sum())
```

```python
print(dataset.select_dtypes(include="object").columns)
```

```python
for i in dataset.select_dtypes(include="object").columns:
    dataset[i].fillna(dataset[i].mode()[0], inplace=True)
```

```python
print(dataset.select_dtypes(include="object").isnull().sum())
```

```python
print(dataset.select_dtypes(include="float64").isnull().sum())
```

```python
print(dataset.select_dtypes(include="float64").columns)
```

```python
for i in dataset.select_dtypes(include="float64").columns:
    dataset[i].fillna(dataset[i].mean(), inplace=True)
```

```python
print(dataset.select_dtypes(include="float64").isnull().sum())
```

```python
print(dataset.isnull().sum())
```

```python
ne_dataset = (dataset[["fuel", "transmission"]])
print(ne_dataset)
```

```python
print(ne_dataset.isnull().sum())
```

```python
pd.get_dummies(ne_dataset).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(ne_dataset).toarray()
```
