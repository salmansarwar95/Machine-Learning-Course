```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt = pd.read_csv("Car details v3.csv")
print(dt.info())
```

```python
print(dt.head(2))
```

```python
print(dt.isnull().sum())
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(dt.isnull())
```

```python
print(dt.select_dtypes(include="object").isnull().sum())
```

```python
print(dt.select_dtypes(include="object").columns)
```

```python
for i in dt.select_dtypes(include="object").columns:
    dt[i].fillna(dt[i].mode()[0], inplace=True)
```

```python
print(dt.select_dtypes(include="object").isnull().sum())
```

```python
print(dt.select_dtypes(include='float64').isnull().sum())
```

```python
print(dt.select_dtypes(include='float64').columns)
```

```python
for i in dt.select_dtypes(include='float64').columns:
    dt[i].fillna(dt[i].mean(), inplace=True)
```

```python
print(dt.select_dtypes(include='float64').isnull().sum())
```

```python
print(dt.isnull().sum())
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(dt.isnull())
```
