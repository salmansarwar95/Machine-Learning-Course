```python
# through the code, mean simple output
print(af["Drivetrain"].mode()[0]) #code for checking mode of any column; [0] use for simple output. without it output comes in series dataset

# code for to fill mode value in the place Nan
af['Drivetrain'].fillna(af["Drivetrain"].mode()[0], inplace=True)
print(af.isnull().sum())

# code for checking mode value by graphical representation
qq = (af["Drivetrain"])
sns.countplot(x=qq, data=af, edgecolor='black')
plt.show()
```

Collecting "object" Dtype columns

```python
# Orignal code for collecting
# print(af.select_dtypes(include="object")) 


print(af.select_dtypes(include="object").isnull().sum())
```

```python
for i in af.select_dtypes(include="object").columns:
    af[i].fillna(af[i].mode()[0], inplace=True)

print(af.isnull().sum())
```

Collecting "float64" Dtype columns

```python
print(af.select_dtypes(include="float64").isnull().sum())
```

```python
print(af.select_dtypes(include="float64").columns)
```

```python
for i in af.select_dtypes(include="float64").columns:
    af[i].fillna(af[i].mean(), inplace=True)

print(af.isnull().sum())
```
