```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Car details v3.csv")
print(df.head(2))
```

```python
print(df.count())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
print(df.isnull().sum().sum())
```

```python
print((df.isnull().sum())/len(df)*100)
print((df.isnull().sum().sum())/len(df)*100)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull())
plt.title("Graphical Representation")
plt.tight_layout()
plt.show()
```

```python
aa = (df["transmission"])
# a = (aa.head(100))
plt.figure(figsize=(10,5))
sns.countplot(x=aa, data=df)
```
