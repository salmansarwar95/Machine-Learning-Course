# Through The IQR Method

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_dataset = pd.read_csv('Iris.csv')
iris_dataset.head(10)
```

```python
iris_dataset.isnull().sum()
```

```python
iris_dataset.describe()
```

```python
sns.boxplot(x = "PetalLengthCm", data=iris_dataset)
plt.show()
```

```python
q1 = iris_dataset["PetalLengthCm"].quantile(0.25)
q3 = iris_dataset["PetalLengthCm"].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
min_rng = q1 - (1.5*IQR)
max_rng = q3 + (1.5*IQR)
```

```python
min_rng, max_rng
```

```python
n_iris_dataset = iris_dataset[iris_dataset["PetalLengthCm"] <= max_rng]
```

```python
sns.boxplot(x = "PetalLengthCm", data=n_iris_dataset)
plt.show()
```
