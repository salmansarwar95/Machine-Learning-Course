```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d_s = pd.read_csv("car data.csv")
d_s.info()
```

```python
d_s.describe()
```

```python
sns.boxplot(x = "Present_Price", data=d_s)
```

### Through IQR

```python
q1 = d_s["Present_Price"].quantile(0.25)
q3 = d_s["Present_Price"].quantile(0.75)
```

```python
IQR = q3-q1
```

```python
mins_range = q1 - (1.5*IQR)
maxs_range = q3 + (1.5*IQR) 
```

```python
mins_range, maxs_range
```

```python
n_d_s = d_s[d_s['Present_Price'] <= maxs_range]
```

```python
sns.boxplot(x = "Present_Price", data=n_d_s)
plt.show()
```

### Through Z-Score

```python
min_rang = d_s["Present_Price"].mean() - (3*d_s["Present_Price"].std())
max_rang = d_s["Present_Price"].mean() + (3*d_s["Present_Price"].std())
```

```python
min_rang, max_rang
```

```python
o_n_d_s = d_s[d_s["Present_Price"] <= max_rang]
```

```python
sns.boxplot(x = "Present_Price", data=o_n_d_s)
plt.show()
```
