```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("iris.csv")
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
sns.countplot(x= "Species", data=dataset, color="pink", edgecolor="black")
plt.tight_layout()
plt.show()
```

```python
new_data = dataset[["Species"]]
pd.get_dummies(new_data)
```

```python
new_data.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
e_new_data = ohe.fit_transform(new_data).toarray()
```

```python
encoded_species = pd.DataFrame(e_new_data, columns=["Species_Iris-setosa", "Species_Iris-versicolor", "Species_Iris-virginica"])
```

```python
DataSet = pd.concat([dataset, encoded_species], axis=1)
DataSet
```

```python
DataSet.drop(["Species"], axis=1, inplace=True)
```

```python
DataSet.drop(["Id"], axis=1, inplace=True)
```
