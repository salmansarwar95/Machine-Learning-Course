```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DF = pd.DataFrame({
    "Study Hours" : [4, 5, 6, 7, 8],
    "Marks" : [50, 60, 70, 80, 90]
})

print(DF)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

stander_scaler = StandardScaler()
stander_scaled = stander_scaler.fit_transform(DF)
print("\nStander_scaler Output:")
print(pd.DataFrame(stander_scaled, columns=["Study Hours", "Marks"]))

minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(DF)
print("\nMinmax_scaler Output:")
print(pd.DataFrame(minmax_scaled, columns=["Study Hours", "Marks"]))

```

```python
from sklearn.model_selection import train_test_split

X = DF[["Study Hours"]]
y = DF[["Marks"]]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=40
)
```

```python
print("\nTraning Data")
print(X_train)

print("\nTesting Data")
print(X_test)
```
