```python
import matplotlib.pyplot as plt
```

```python
Students = ["Salman", "Nadeem", "Maaz", "Zohaib", "Sudaif"]
Departmants  = ["Ai", "SE", "CS", "SC", "IT"]
Enrolments = [1000, 1152, 2000, 1800, 1300]
Roll_No = [1002, 1022, 1015, 1009, 1027]
```

```python
plt.scatter(Students, Departmants, color = "red", edgecolors = "black")
plt.title("New-Enrolments")
plt.xlabel("Names")
plt.ylabel("Departments")
plt.show()
```

```python
plt.scatter(Students, Roll_No, color = "red")
plt.title("Student's Roll-No")
plt.xlabel("Names")
plt.ylabel("Roll-No")
plt.show()
```

```python
plt.scatter(Departmants, Enrolments, color = "brown")
# plt.plot(Departmants, Enrolments, color = "brown")
plt.title("Numbers of New-Enrolments")
plt.xlabel("Departmant")
plt.ylabel("Enrolments")
plt.show()
```
