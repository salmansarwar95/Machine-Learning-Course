### Used for category comparison

```python
import matplotlib.pyplot as plt
```

```python
Age = [22, 25, 26, 28, 32]
Hieght = [5.5, 5.1, 6, 6.2, 5]
```

```python
plt.bar(Age, Hieght, color = "orange", edgecolor = "black")
plt.title("Hieght wrt Age")
plt.xlabel("Age")
plt.ylabel("Hieght(feet)")
plt.show()
```

```python
year = [2012, 2013, 2014, 2015, 2016]
profit = [22, 34, 12, 45, 30]
```

```python
plt.bar(year, profit, color = "red")
plt.title("Profit per Annum")
plt.xlabel("Year")
plt.ylabel("Profit (Million)")
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
Designation = ["Compaodar", "Cooker man", "Nurse", "Pharmacist", "Doctor"]
salries = [33300, 20000, 50000, 90000, 70000]
```

```python
plt.bar(Designation, salries, color = "pink", edgecolor = "black")
plt.title("Hospital Staff's Salries")
plt.xlabel("Designation")
plt.ylabel("Salries")
plt.grid(True)
plt.show()
```

```python
plt.plot(Designation, salries, color = "blue", marker = "*", linestyle = ":")
plt.title("Hospital Staff's Salries")
plt.xlabel("Designation")
plt.ylabel("Salries")
plt.grid()
plt.show()
```
