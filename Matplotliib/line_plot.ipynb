### Line Plot(Use for times trend)

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [10, 20, 30]
```

```python
plt.plot(x, y, color = "Red", marker = "o", linestyle = "-." )
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x = ["Computer", "Math", "English", "Urdu", "Islamiat"]
y = [70, 98, 95, 77, 99]
```

```python
plt.plot(x, y, color = "Brown", marker = "*")
plt.title("Grades")
plt.xlabel("------ Subjects ------>")
plt.ylabel("------ Markes ------>")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [2020, 2021, 2022, 2023, 2024, 2025]
y = [555.5, 800, 400, 700, 900, 600.5]

plt.plot(x, y, color = "red", marker = "*", linewidth = 2)
plt.title("\nAnnual Health Budget\n")
plt.xlabel("\nYears\n--------->\n")
plt.ylabel("\nBudget\n--------------------->\n(Billion Dollars)\n")
plt.grid(True)
plt.tight_layout()
plt.show()
```
