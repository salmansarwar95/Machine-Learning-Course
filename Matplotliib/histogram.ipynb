### For frequently distribution

```python
import matplotlib.pyplot as plt
```

```python
ages = [22, 25, 29, 21, 25, 24, 23, 29, 26, 28]

plt.figure(figsize=(10,5))
plt.hist(ages, bins=5, color="green", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt


Names = ["Salman", "Saqlain", "Zeeshan", "Jamshad", "Jamshad", "Asad", "Zeeshan", "Jawad", "Zain", "Salman"]

plt.hist(Names, bins=6, color = "pink", edgecolor = "black")
plt.title("Names Ocurance", color = "red", fontsize = 20)
plt.xlabel("Names", fontsize = 12)
plt.ylabel("Ocurance", fontsize = 12)
plt.show()
```
