```python
import matplotlib.pyplot as plt
categories = ['a','b', 'c']
values = [1,2,3]
x = ['Akbar', 'Ali', 'Bilal', 'Zain']
y = [1800, 2000, 1500, 2200]
plt.figure(figsize=(15,5))

plt.subplot(1, 3, 1)
plt.bar(categories, values, color = 'yellow', edgecolor = "black")
plt.title("Bar Chart")

plt.subplot(1, 3, 2)
plt.pie(y, labels=x, autopct ="%1.1f%%", startangle=90)
plt.title("Line Chart")

plt.subplot(1, 3, 3)
plt.plot(x, y, color = "red", marker = "o")
plt.title("Line Chart")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

Name = ["Hassan", "Ali", "Majid","Asim", "Zain", "Asad"]
Salary = [30000, 32000, 28000, 25000, 35000, 31000]
Age = [20, 22, 18, 25, 24, 21]

plt.figure(figsize=(15,5))

plt.subplot(1, 4, 1)
plt.bar(Name, Salary, color = "pink", edgecolor = "black")
plt.title("Salary PKG")
plt.xlabel('Names')
plt.ylabel('Salaries')


plt.subplot(1, 4, 2)
plt.plot(Name, Age, color = 'green', marker = "*")
plt.title("Worker Ages")
plt.xlabel("Names")
plt.ylabel("Ages")
plt.grid(True)

plt.subplot(1, 4, 3)
plt.scatter(Name, Salary, color = "purple")
plt.title("Pointout Salaries")
plt.xlabel('Names')
plt.ylabel("Salaries")

plt.subplot(1, 4, 4)
plt.pie(Salary, labels=Name, autopct='%1.1f%%', startangle=90)
plt.title("Salaray Percentage")
plt.axis("Equal")

plt.tight_layout()
plt.show()
```
