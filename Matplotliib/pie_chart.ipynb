```python
import matplotlib.pyplot as plt
```

```python
subject = ['Math', 'English', 'Urdu', 'Islamiyat']
percentage = [99, 92, 80, 95]
```

```python
plt.pie(percentage, labels=subject, autopct='%1.1f%%', startangle=90 )
plt.axis("equal")
plt.title("Grades Percentage in Final Result", fontsize = 15)
plt.show()
```
