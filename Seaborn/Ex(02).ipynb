```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Data = sns.load_dataset("tips")
```

```python
print(Data.count())
```

```python
print(Data.head(20))
```

```python
SZ = (Data["size"].head(10))
print(SZ)

plt.figure(figsize=(5,5))
sns.countplot(x = SZ, color="pink", edgecolor = "black")
plt.title("Total Number of Bills")
plt.xlabel("Bill Size")
plt.tight_layout()
plt.show()
```

```python
Smokers = (Data['smoker'].head(100))
print (Smokers)

plt.figure(figsize=(5,5))
sns.countplot(x = Smokers, color="green", edgecolor= "black")
plt.title('Smoker numbers')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
dd = (Data.head(15))
print (dd)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = "total_bill", y = "tip", data=dd, color = "purple", edgecolor = "black", )
plt.title("Total Bill wrt Size")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
siz = (Data["size"])
aa = (siz.head(25))
print(aa)

plt.figure(figsize=(5,5))
sns.countplot(x = aa, data=Data, color = "gray", edgecolor  = "black")
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
Tips = (Data["tip"])
Tip =  (Tips.head(25))
print (Tip)

plt.figure(figsize=(10,5))
sns.countplot(x = Tip, data=Data, color = 'orange', edgecolor = "black")
plt.title('Tips')
plt.tight_layout()
plt.grid(True)
plt.show()
```
