### Creating Arrays from Scratch

```python
import numpy as np
zero = np.zeros((2, 2))
print ("Zeros: \n", zero)

one = np.ones((3, 3))
print ("Ones: \n", one)

full = np.full((2, 2), 6)
print ("Full: \n", full)

random = np.random.random((3,3))
print ("Random: \n", random)

sequence = np.arange(0, 10, 2)
print("Squence: \n", sequence)
```
