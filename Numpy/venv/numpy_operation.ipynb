### Numpy Operations

Array Indexing


```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print ("Indexing:\n", arr[1:4])
```

```python
import numpy as np
matrix = np.array([[11, 22, 33], #also callde 2d array
[44, 55, 66],
[77, 88, 99]])
print ("Matrix: \n", matrix[:, 2]) #leave row and focus on the colums
print ("Matrix: \n", matrix[2]) #leave colums and focus on the row
print ("Matrix: \n", matrix[1,2]) #focus on the mentioned line single item

```
